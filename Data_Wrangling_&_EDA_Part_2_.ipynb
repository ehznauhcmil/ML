## MatPlotLib and Seaborn

Both Matplotlib and Seaborn are extensively used in data analysis, machine learning, and scientific research. They offer a wide range of visualization options and can be integrated with other Python libraries to enhance data exploration, presentation, and communication of insights. Whether you need basic or advanced plotting capabilities, Matplotlib and Seaborn provide the tools to create** visually appealing and insightful visualizations.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Matplotlib:
Matplotlib is a popular data visualization library for creating static, animated, and interactive visualizations in Python. It provides a wide range of plotting functionalities and allows you to create various types of plots, such as line plots, scatter plots, bar charts, histograms, and more. Matplotlib provides fine-grained control over every aspect of a plot, enabling customization of colors, labels, markers, and axes. It is a versatile library that is widely used for data exploration, analysis, and presentation.

Seaborn:
Seaborn is a powerful statistical data visualization library built on top of Matplotlib. It provides a high-level interface for creating informative and visually appealing statistical graphics. Seaborn simplifies the process of creating complex visualizations by offering several built-in themes and color palettes. It also provides a set of specialized plot types, including box plots, violin plots, heatmaps, and pair plots, that are commonly used in statistical analysis. Seaborn's main strength lies in its ability to easily create aesthetically pleasing and informative visualizations with minimal code.

```python
data = pd.read_csv('part2.csv')

data
```

##Univariate Analysis
Univariate analysis focuses on exploring and understanding a single variable at a time.


Histogram

A histogram is a graphical representation of the distribution of a continuous variable. It displays the frequency of values falling within different bins or intervals. The height of each bar represents the frequency or count of values in that bin.

To create a histogram, we can use the matplotlib library's hist function.

```python
plt.hist(data['Age'], bins=10, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

Here, we create a histogram to visualize the distribution of the 'Age' variable. The x-axis represents the age values, and the y-axis represents the frequency. The histogram allows us to observe the distribution of ages and identify any patterns or outliers

Interpretation: The age distribution appears to be relatively evenly distributed, with the majority of students falling between the ages of 17 and 20.

Box Plot

A box plot, also known as a box-and-whisker plot, displays the summary statistics of a continuous variable, including the minimum, maximum, median, and quartiles. It provides insights into the central tendency, spread, and skewness of the data.

For a box plot, we can utilize matplotlib or seaborn.

Using matplotlib:

```python
plt.boxplot(data['ExamScore'])
plt.xlabel('Exam Score')
plt.title('Exam Score Distribution')
plt.show()
```

Here, we create a box plot to visualize the distribution of 'ExamScore'. The box represents the interquartile range (IQR), the line inside the box represents the median, and the whiskers represent the range of values within 1.5 times the IQR. Outliers are shown as individual points outside the whiskers.

Interpretation: The exam scores show a relatively symmetrical distribution with a median score around 85. There are a few outliers on the lower end of the scores.

Using seaborn:

```python
sns.boxplot(data=data, x='Gender', y='StudyHours')
plt.xlabel('Gender')
plt.ylabel('Study Hours')
plt.title('Study Hours by Gender')
plt.show()
```

Frequency Table

A frequency table summarizes the count or frequency of each category or value in a categorical variable. It provides an overview of the distribution of categorical data.

To generate a frequency table, we can utilize pandas's value_counts method.

Interpretation: The frequency table reveals that there are an equal number of male and female students in the dataset.

```python
frequency_table = data['Gender'].value_counts()
print(frequency_table)
```

Here, we create a frequency table to show the count of each gender category. The frequency table allows us to observe the distribution of genders and determine the most common gender category.

##Bivariate Analysis

Bivariate analysis explores the relationship between two variables simultaneously.



Scatter Plot

A scatter plot displays the relationship between two continuous variables. Each data point is represented as a dot on the plot, with the x-axis representing one variable and the y-axis representing the other variable. Scatter plots are useful for identifying patterns, trends, and the strength of the relationship between the variables.

To create a scatter plot, we can use the matplotlib library's scatter function.

```python
plt.scatter(data['StudyHours'], data['ExamScore'])
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.title('Relationship between Study Hours and Exam Score')
plt.show()
```

Here, we create a scatter plot to visualize the relationship between 'StudyHours' and 'ExamScore'. Each data point represents a student's study hours and exam score. The scatter plot allows us to observe if there is a linear or nonlinear relationship between study hours and exam scores.

Interpretation: The scatter plot suggests a positive relationship between study hours and exam scores. As the number of study hours increases, the exam scores tend to be higher as well.


Bar Chart

A bar chart represents the distribution or comparison of a categorical variable across different categories. It uses rectangular bars to represent the count, frequency, or proportion of each category.

For a bar chart, we can use matplotlib or seaborn.

Using matplotlib:

```python
plt.bar(data['Gender'], data['ExamScore'])
plt.xlabel('Gender')
plt.ylabel('Exam Score')
plt.title('Exam Scores by Gender')
plt.show()
```

Here, we create a bar chart to compare the average 'ExamScore' across different 'Gender' categories. The height of each bar represents the average exam score for each gender category. The bar chart allows us to compare the exam scores between genders.

Interpretation: The bar chart shows that, on average, females have slightly higher exam scores compared to males.

Using seaborn:

```python
sns.barplot(data=data, x='Gender', y='ExamScore')
plt.xlabel('Gender')
plt.ylabel('Exam Score')
plt.title('Exam Scores by Gender')
plt.show()
```

Correlation Matrix

A correlation matrix shows the pairwise correlations between multiple variables. It is often represented as a heatmap, where each cell represents the correlation coefficient between two variables. Correlation coefficients range from -1 to 1, indicating the strength and direction of the relationship (negative, positive, or no correlation) between variables.


To calculate and visualize the correlation matrix, we can use pandas and seaborn.

```python
# Calculate the correlation matrix
correlation_matrix = data.corr()
# Create a heatmap using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='RdYlBu', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

Here, we calculate the correlation matrix for all numerical variables in the dataset. Then, we create a heatmap to visualize the correlation matrix. The heatmap uses a color scale to represent the correlation values, with warmer colors indicating positive correlations and cooler colors indicating negative correlations. The correlation matrix heatmap allows us to identify relationships and dependencies between variables.

Interpretation: The heatmap indicates a strong positive correlation between study hours and exam scores, suggesting that students who study more tend to achieve higher scores.
